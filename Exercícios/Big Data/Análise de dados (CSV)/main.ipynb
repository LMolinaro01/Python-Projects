```python
#%pip install pandas
import pandas as pd
from IPython.display import display, HTML 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('global_video_game_sales.csv') #Essa variável "df" agora contém um DataFrame, que é uma estrutura de dados do pandas que representa uma tabela bidimensional, similar a uma planilha ou tabela de banco de dados.
df.head() #Quando você executa esse comando, o pandas mostra as primeiras 5 linhas do seu DataFrame por padrão. Isso é útil quando você deseja obter uma visão rápida dos dados para verificar sua estrutura, tipos de dados e valores nas primeiras linhas.
```

```python
query_result = df[df['Global_Sales'] > 30]
print(query_result)
```

```python
#Exibição dos Primeiros 20 Jogos na Plataforma DS (Sem Ordem Cronológica)

query_result = df[df['Platform'] == "DS"]
limited_query_result = query_result[['Name', 'Platform']].head(20)
display(HTML(limited_query_result.to_html(border=1)))
```

```python
# Número de Jogos por Ano (Gênero: Sports, Publisher: Nintendo, Plataforma: Wii

query_result = df[(df['Genre'] == "Sports") & (df['Publisher'] == "Nintendo") & (df['Platform'] == "Wii") ]
limited_query_result = query_result.head(20)
display(HTML(query_result.to_html(border=1)))

year_counts = query_result['Year'].value_counts().sort_index()

# Gerar gráfico de barras
plt.figure(figsize=(10,6))
year_counts.plot(kind='bar', color='lightgreen')
plt.title('Número de Jogos por Ano (Gênero: Sports, Publisher: Nintendo, Plataforma: Wii)', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Número de Jogos', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Análise das Vendas Globais por Ano para Jogos de Esporte da Nintendo no Wi


query_result = df[(df['Genre'] == "Sports") & (df['Publisher'] == "Nintendo") & (df['Platform'] == "Wii")]

# Limitar para as primeiras 20 linhas
limited_query_result = query_result.head(20)

# Exibir a tabela com o grid no Jupyter Notebook
display(HTML(limited_query_result.to_html(border=1)))

# Somar as vendas globais por ano
sales_by_year = query_result.groupby('Year')['Global_Sales'].sum().sort_index()

# Gerar gráfico de barras para vendas globais por ano
plt.figure(figsize=(10,6))
sales_by_year.plot(kind='bar', color='coral')
plt.title('Vendas Globais por Ano (Gênero: Sports, Publisher: Nintendo, Plataforma: Wii)', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Vendas Globais (milhões)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Distribuição do Número de Jogos por Plataforma (Gênero: Action, Publisher: Nintendo, Ano <= 2000)


query_result = df[(df['Genre'] == "Action") & (df['Year'] <= 2000) & (df['Publisher'] == "Nintendo")]
limited_query_result = query_result[['Name', 'Platform']].head(20)


# Contar o número de jogos por plataforma
platform_counts = query_result['Platform'].value_counts()

# Gerar gráfico de barras
plt.figure(figsize=(10,6))
platform_counts.plot(kind='bar', color='skyblue')
plt.title('Número de Jogos por Plataforma (Gênero: Action, Publisher: Nintendo, Ano <= 2000)', fontsize=14)
plt.xlabel('Plataforma', fontsize=12)
plt.ylabel('Número de Jogos', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

display(HTML(query_result.to_html(border=1)))
```
