```python
import pandas as pd
from functools import reduce

data = pd.read_json("dados_video_game.json")
```

```python
data.head()
```

```python
data.loc[(data["Year"] > 2010) & (data["Year"] < 2015) & (data["Platform"] == "PSThree")]
```

```python
data["Global_Sales"] = data["Global_Sales"].astype(float)

data["Altíssima_Venda"] = data["Global_Sales"].apply(lambda valor: valor > 20)
data["Alta_Venda"] = data["Global_Sales"].apply(lambda valor: valor > 10)
data["Boa_Venda"] = data["Global_Sales"].apply(lambda valor: valor > 1)
data["Média_Venda"] = data["Global_Sales"].apply(lambda valor: valor < 1)
data["Baixa_Venda"] = data["Global_Sales"].apply(lambda valor: valor < 0.1)
data["Vendas_Unidades"] = data["Global_Sales"].astype(float).apply(lambda x: int(x * 1_000_000))
data["Vendas_Formatadas"] = data["Vendas_Unidades"].apply(lambda x: f"{x:,}".replace(",", "."))

print(data[["Name", "Global_Sales", "Alta_Venda"]].head())
```

```python
data.loc[(data["Global_Sales"] < 0.1) & (data["Platform"] == "PSThree")]
```

```python
# 1. Sum - Soma os dados de uma coluna
total_na_sales = data['NA_Sales'].sum()
print(f"Soma das vendas na América do Norte: {total_na_sales}")
```

```python
# 2. Filter - Filtra valores maiores que 10 em 'NA_Sales'
maior_que_10 = data['NA_Sales'].apply(lambda x: x > 10)
num_filtrados = data[maior_que_10]
print(f"Jogos com vendas maiores que 10 milhões na América do Norte:\n{num_filtrados}")
```

```python
# 3. Map - Multiplicar cada valor em 'Global_Sales' por 3
data['Global_Sales_triplicado'] = data['Global_Sales'].map(lambda x: x * 3)
print(f"DataFrame com 'Global_Sales' triplicados:\n{data}")
```

```python
# 4. Reduce - Somar todas as vendas globais
total_global_sales = reduce(lambda x, y: x + y, data['Global_Sales'])
print(f"Soma total de 'Global_Sales' usando reduce: {total_global_sales}")
```
